```python
# Iterative Deepening A* (IDA*) Algorithm

# Heuristic Function
def heuristic(n):
    H_dist = {
        'A': 11,
        'B': 6,
        'C': 99,
        'D': 1,
        'E': 7,
        'G': 0,
    }
    return H_dist[n]

# Graph Representation
Graph_nodes = {
    'A': [('B', 2), ('E', 3)],
    'B': [('A', 2), ('C', 1), ('G', 9)],
    'C': [('B', 1)],
    'D': [('E', 6), ('G', 1)],
    'E': [('A', 3), ('D', 6)],
    'G': [('B', 9), ('D', 1)]
}

# Function to get neighbors
def get_neighbors(v):
    if v in Graph_nodes:
        return Graph_nodes[v]
    else:
        return []

# Recursive Search Function
def search(path, g, threshold, goal):

    node = path[-1]

    # f(n) = g(n) + h(n)
    f = g + heuristic(node)

    # If f-cost exceeds threshold
    if f > threshold:
        return f

    # Goal Test
    if node == goal:
        return "FOUND"

    minimum = float('inf')

    # Explore neighbors
    for (neighbor, weight) in get_neighbors(node):

        if neighbor not in path:

            path.append(neighbor)

            temp = search(path,
                          g + weight,
                          threshold,
                          goal)

            # Goal found
            if temp == "FOUND":
                return "FOUND"

            # Minimum threshold update
            if temp < minimum:
                minimum = temp

            path.pop()

    return minimum

# IDA* Main Algorithm
def ida_star(start_node, stop_node):

    # Initial threshold
    threshold = heuristic(start_node)

    # Initial path
    path = [start_node]

    while True:

        temp = search(path,
                      0,
                      threshold,
                      stop_node)

        # Goal Found
        if temp == "FOUND":
            print("Path found:", path)
            return path

        # No path exists
        if temp == float('inf'):
            print("Path does not exist!")
            return None

        # Update threshold
        threshold = temp

# Driver Code
ida_star('A', 'G')
```
